```python
import numpy as np
```

```python
data.dropna(subset=['date_added','rating'], inplace=True)
data.isna().sum()
```

```python
data["date_added"]=pd.to_datetime(data["date_added"])
data.head()
```

```python
data.isnull().sum()
```

```python
data.head()
```

```python
from sklearn.impute import SimpleImputer
```

```python
data["duration_minutes"].describe()
```

```python
imp_num=SimpleImputer(missing_values=np.NaN,strategy='median')
```

```python
missing_num_col=["duration_minutes"]
```

```python
data[missing_num_col]=imp_num.fit_transform(data[missing_num_col])
```

```python
data.head()
```

```python
data["duration_seasons"].describe()
```

```python
imp_cat=SimpleImputer(missing_values=np.NaN,strategy='most_frequent')
```

```python
missing_cat_col=["duration_seasons"]
```

```python
data[missing_cat_col]=imp_cat.fit_transform(data[missing_cat_col])
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
data["release_year"]=data["release_year"].astype(int)
```
